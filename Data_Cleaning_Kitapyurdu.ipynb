```python
import pandas as pd
import numpy as np
import warnings
pd.options.display.max_colwidth = 50
pd.options.display.max_columns = None
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Books_Scrap.csv')
```

```python
df.sample(10)
```

```python
df.rename(columns={'0':'name', '1':'writer', '2':'publisher', '3':'about', '4':'informations', '5':'rate', '6':'comment', '7':'read_status', '8':'sales', '9':'price'}, inplace=True)
```

```python
col_list = []
for i in df.informations:
    col = i.replace(':','').split('\n')[::4]
    for ii in col:
        if ii not in col_list:
            col_list.append(ii)
for i in col_list:
    df[i] = np.nan
for i in range(len(df)):
    title = df.informations[i].replace(':','').split('\n')[::4]
    index = df.informations[i].replace(':','').split('\n')[1::4]
    for ii in range(len(title)):
        df[title[ii]][i] = index[ii]
df.drop(columns='informations', axis=1, inplace=True)
```

```python
df.comment = df.comment.astype('str')
df.comment = df.comment.apply(lambda x: x.replace('.0','').replace('.',''))
df.comment = df.comment.astype('int')
```

```python
df['Okuyacağım'] = np.nan
df['Okuyorum'] = np.nan
df['Okudum'] = np.nan

for i in range(len(df)):
    locs = df.read_status[i].strip().split()
    df[locs[0]][i] = locs[1]
    df[locs[2]][i] = locs[3]
    df[locs[4]][i] = locs[5]
df.drop(columns='read_status', axis=1, inplace=True)
```

```python
df.sales = df.sales.astype('str')
df.sales = df.sales.apply(lambda x: x.replace('.0','').replace('.',''))
df.sales = df.sales.astype('int')
```

```python
df.price = df.price.apply(lambda x: x.strip().replace(',','.'))
df.price = df.price.astype('float')
```

```python
df.rename(columns={'Çevirmen':'translator','Yayın Tarihi':'publish_date','Orijinal Adı':'original_name','ISBN':'isbn','Dil':'language','Sayfa Sayısı':'page','Cilt Tipi':'volume_type','Kağıt Cinsi':'paper_type','Boyut':'size','Okuyacağım':'will_read','Okuyorum':'reading','Okudum':'have_read'}, inplace=True)
```

```python
df.drop(columns=['Kapak','Editor','Çizer','Diğer Özellik','Derleyici'], inplace=True)
```

```python
for i in df[df.translator.isnull() == True].index:
    df.translator[i] = df.writer[i]
for i in df[df.original_name.isnull() == True].index:
    df.original_name[i] = df.name[i]
df['size'].fillna('13.5 x 21 cm', inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isbn = df.isbn.astype('int64')
```

```python
df.page = df.page.astype('int')
```

```python
df.will_read = df.will_read.apply(lambda x: x.replace('.',''))
df.will_read = df.will_read.astype('int')
df.reading = df.reading.apply(lambda x: x.replace('.',''))
df.reading = df.reading.astype('int')
df.have_read = df.have_read.apply(lambda x: x.replace('.',''))
df.have_read = df.have_read.astype('int')
```
