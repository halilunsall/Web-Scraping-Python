```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import regex as re
```

```python
product_name = []
product_rate = []
product_comment = []
product_price = []
```

```python
url = 'https://www.amazon.com.tr/gp/bestsellers/electronics/ref=zg-bs_electronics_dw_sml'
req = requests.get(url, {'user-agent': 'YOUR-HEADER'})
soup = BeautifulSoup(req.content,'lxml')
products = soup.find_all('div', {'class':'a-column a-span12 a-text-center _cDEzb_grid-column_2hIsc'})

for product in products:
    product_name.append(product.find('div',{'class':'_cDEzb_p13n-sc-css-line-clamp-3_g3dy1'}).text)
    product_rate.append(re.search('(\d,\d).(\d.{1,5})', product.find('div',{'class':'a-row'}).text)[1])
    product_comment.append(re.search('(\d,\d).(\d.{1,5})', product.find('div',{'class':'a-row'}).text)[2])
    product_link = 'https://www.amazon.com.tr/'+product.find('a',{'class':'a-link-normal'}).get('href')
    product_in = requests.get(product_link, headers = {'user-agent': 'YOUR-HEADER'})
    product_details = BeautifulSoup(product_in.content,'lxml')
    product_price.append(product_details.find('span', {'class':'a-offscreen'}).text)
```

```python
pd.DataFrame({'product_name':product_name, 'product_rate':product_rate, 'product_comment':product_comment, 'product_price':product_price})
```
