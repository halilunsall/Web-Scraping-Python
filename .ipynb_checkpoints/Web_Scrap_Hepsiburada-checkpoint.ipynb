```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import numpy as np
```

```python
url = 'https://www.hepsiburada.com/oyuncu-laptoplari-c-95583'
head = {"User-Agent": "YOUR-HEADER"}
r = requests.get(url,headers=head)
soup = BeautifulSoup(r.content, 'lxml')
```

```python
part = soup.find('div', {'class':'productListContent-pXUkO4iHa51o_17CBibU'})
```

```python
part1 = part.find('ul', {'class':'productListContent-frGrtf5XrVXRwJ05HUfU productListContent-rEYj2_8SETJUeqNhyzSm', 'id':'1'})
```

```python
part2 = part1.find_all('li', {'class':'productListContent-zAP0Y5msy8OHn5z7T_K_'})
```

```python
name = []
price = []
rate = []
comment = []
techs = []
seller = []
seller_rate = []
```

```python
for detail in part2:
    link = 'https://www.hepsiburada.com/'+detail.a.get('href')
    r1= requests.get(link, headers={'user-agent': 'YOUR-HEADER'})
    soup1 = BeautifulSoup(r1.content,'lxml')
    name.append(soup1.find('h1',{'id':'product-name'}).text.strip())
    price.append(soup1.find('del',{'id':'originalPrice'}).text)
    rate.append(soup1.find('span', {'class':'rating-star'}).text.strip())
    comment.append(soup1.find('div', {'id':'comments-container'}).span.text.strip())
    techs.append(soup1.find('table', {'class':'data-list tech-spec'}).tbody.text.strip())
    seller.append(soup1.find('div',{'class':'seller-container'}).span.a.text.strip())
    try:
        seller_rate.append(soup1.find('span',{'class':'merchant-rating lv-5'}).text.strip())
    except:
        seller_rate.append(np.nan)
```

```python
df = pd.DataFrame({'product_name':name, 'product_price': price, 'product_rate':rate, 'product_comment':comment, 'product_prop':techs, 'product_seller':seller, 'product_seller_rate':seller_rate})
```
