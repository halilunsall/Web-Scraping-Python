```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'

res = requests.get(base_url.format(1))
soup = BeautifulSoup(res.text, 'lxml')
books = soup.find('ol',{'class':'row'}).find_all('li')

book_name = []
book_price = []
book_rate = []

for item in books:
    book_name.append(item.select('a')[1]['title'])
    book_price.append(item.find('p', {'class':'price_color'}).text.lstrip('Â£'))
    i_rate = item.select('p')[0]['class'][1]
    if i_rate == 'One':
        book_rate.append('1')
    elif i_rate == 'Two':
        book_rate.append('2')
    elif i_rate == 'Three':
        book_rate.append('3')
    elif i_rate == 'Four':
        book_rate.append('4')
    elif i_rate == 'Five':
        book_rate.append('5')
```

```python
book_name = []
book_price = []
book_rate = []

for num_page in range(1,51):
    scrape_url = base_url.format(num_page)
    res = requests.get(scrape_url)
    
    soup = BeautifulSoup(res.text, 'lxml')
    books = soup.find('ol',{'class':'row'}).find_all('li')


    for item in books:
        book_name.append(item.select('a')[1]['title'])
        book_price.append(item.find('p', {'class':'price_color'}).text.lstrip('Â£'))
        i_rate = item.select('p')[0]['class'][1]
        if i_rate == 'One':
            book_rate.append('1')
        elif i_rate == 'Two':
            book_rate.append('2')
        elif i_rate == 'Three':
            book_rate.append('3')
        elif i_rate == 'Four':
            book_rate.append('4')
        elif i_rate == 'Five':
            book_rate.append('5')
```

```python
df = pd.DataFrame({'Name': book_name, 'Price': book_price, 'Rate':book_rate})
df
```

```python
df.to_csv('Books_Scrap.csv')
```
