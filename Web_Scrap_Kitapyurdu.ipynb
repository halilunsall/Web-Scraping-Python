```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import regex
import numpy as np
```

```python
url = 'https://www.kitapyurdu.com/index.php?route=product/best_sellers&page={}&list_id=16&filter_in_stock=1&filter_in_stock=1&limit=100'
books = []

for num in range(1,6):
    new_url = url.format(num)
    r = requests.get(new_url)
    soap = BeautifulSoup(r.content, 'lxml')
    products = soap.find_all('div', {'class':'product-cr'})
    for i in products:
        try:
            book_url = i.a.get('href')
            book_r = requests.get(book_url)
            book_soup = BeautifulSoup(book_r.content, 'lxml')
            a_name = book_soup.find('div', {'class':'pr_header'}).text.strip()
            a_writer = book_soup.find('a', {'class':'pr_producers__link'}).text.strip()
            a_publisher = book_soup.find('div', {'class':'pr_producers__publisher'}).text.strip()
            a_about = book_soup.find('span', {"class":"info__text"}).text
            a_inf = book_soup.find('div', {'class':'attributes'}).text.strip()
            a_rate = len(book_soup.find_all('i', {"class":"sprite__product icon__star-big icon__star-big--selected"}))
            a_comment = book_soup.find('p', {"class":"pr__htabs-review-text"}).text.split()[1]
            a_read_status = book_soup.find('ul', {"class":"readlists"}).text
            a_sales = regex.search('\d+.\d+', book_soup.find('div', {"class":"purchase-info"}).text)[0]
            a_price = book_soup.find('div', {"class":"price__item"}).text
        except:
            print('NoneType')
        
        books.append([a_name, a_writer, a_publisher, a_about,a_inf, a_rate, a_comment, a_read_status, a_sales, a_price])
```

```python
df = pd.DataFrame(books)
```
